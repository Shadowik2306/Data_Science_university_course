### 7. Разобраться с основными возможностями библиотеки Pandas. Необходимо уметь выполнять основные операции по работе с данной библиотекой:

#### 7.1. Загрузка и сохранение данных

```python
import pandas as pd
from pathlib import Path

dataset_path = Path('../datasets/Stores.csv')
df = pd.read_csv(dataset_path, index_col="Store ID ")
```

#### 7.2. Получение сведений о датафрейме с данными

```python
df.info()
```

#### 7.3. Получение сведений о колонках датафрейма;

```python
print(df.describe().transpose())
```

#### 7.4. Вывод отельных строки и столбцов из датафрейма;

```python
print(df["Store_Area"]) # One column
```

```python
print(df[["Store_Area", "Daily_Customer_Count"]]) # Few Columns
```

```python
print(df.loc[100]) # by selected id
```

```python
print(df.iloc[100]) # by real row 
```

```python
print(df.loc[100:200])
```

```python
print(df.loc[100:200, ["Store_Area", "Daily_Customer_Count"]])
```

```python
df.iloc[3:5, 0:3]
```

#### 7.4. Группировка и агрегация данных в датафрейме

```python
print(df.groupby(["Items_Available"]).size().reset_index(name="Count").head(100))
```

#### 7.4. Cортировка данных в датафрейме;

```python
print(df.groupby(["Items_Available"]).size().reset_index(name="Count").sort_values("Count", ascending=False).head(100))
```

#### 7.5. Удаление строк/столбцов

```python
print(df.drop(["Items_Available", "Store_Sales"], axis=1)) # Drop col
```

```python
print(df.drop([1,3,5,100], axis=0)) # Drop col
```

#### 7.6. Cоздание новых столбцов на основе данных из существующих столбцов датафрейма

```python
print(df[["Items_Available", "Daily_Customer_Count"]].groupby(["Daily_Customer_Count"])
      .aggregate(["min", "max"]))
```

#### 7.7. Удаление строк с пустыми значениями;

```python
print(df.dropna()) # No NaN
```

#### 7.8. Заполнение пустых значений на основе существующих данных.

```python
print(df.fillna(value=5)) # No NaN
```

### 8.Разобраться с основными возможностями библиотек Pandas и Matplotlib для визуализации данных из датафрейма:

#### 8.1. Линейная диаграмма (plot);

```python
df.drop("Store_Sales", axis=1).plot()
```

#### 8.2. Cтолбчатая диаграмма (bar);

```python
copy_df = df.copy()
copy_df["Store_Sales"] //= 50
copy_df[["Store_Area", "Store_Sales"]].sort_values("Store_Area").head(50).plot(kind='bar', figsize=(10, 6))
```

#### 8.3. Гистограмма (hist)

```python
df.plot(kind='hist', bins=20, alpha=0.7)
```

```python
df.boxplot(column="Items_Available", by="Store_Area")
```

#### 8.4. Диаграммы с областями (area);

```python
df.plot(kind='area', alpha=0.7)
```

#### 8.5.Диаграмма рассеяния (scatter);

```python
df.plot(kind='scatter', x='Store_Area', y='Store_Sales', title='Scatter Plot')
```

#### 8.6. Круговая диаграмма (pie).

```python
df[['Items_Available']].sort_values('Items_Available').head(5).plot.pie(subplots=True)
```
